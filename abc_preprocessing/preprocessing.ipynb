```python
from glob import glob
import re
```

```python
folders=glob("*", recursive = True)
print(folders)
```

```python
file="./abc/Louis Armstrong/Louis Armstrong - What A Wonderful World.abc"

with open(file) as f:
    originalFile = f.readlines()    
    
for row in originalFile:
    print(row)       
    
generatedFile=[]
```

```python
type(originalFile)
```

```python
for row in originalFile:
    if("X:" in row):
        continue
    elif("T:" in row):
        continue
    elif("M:" in row):
        continue  
    elif("L:" in row):
        continue        
    elif("Q:" in row):
        continue   
    elif("K:" in row): 
        continue 
    elif("V:" in row): 
        continue     
    elif("%%" in row): 
        continue   
    elif("|" in row):
        row=row.replace('|','')
        generatedFile.append(row)    
    else:
        generatedFile.append(row)
```

```python
for row in generatedFile:
    print(row) 
```

```python
createABC = open("processed_Louis Armstrong - What A Wonderful World.abc", "w")
for row in generatedFile:
    createABC.write(row)
createABC.close()
```
